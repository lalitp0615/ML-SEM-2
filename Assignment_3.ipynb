```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Month': ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct'],
    'Product_A_Sales': [75, 78, 85, 88, 90, 92, 95, 96, 98, 100],
    'Product_B_Sales': [30, 55, 58, 60, 75, 78, 79, 82, 90, 97],
    'Production_A': [78, 85, 66, 72, 90, 88, 75, 69, 95, 80],
    'Production_B': [70, 80, 75, 78, 85, 90, 88, 82, 92, 95],
    'Region': ['North','South','East','West','North','South','East','West','North','South']
}

df = pd.DataFrame(data)
df



```

SCATTER PLOT

```python
plt.scatter(df['Month'], df['Product_A_Sales'], color='blue', label='Product A')
plt.scatter(df['Month'], df['Product_B_Sales'], color='red', label='Product B')

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Scatter Plot: Product A vs Product B Sales')
plt.legend()
plt.show()
```

BAR PLOT

```python
x = np.arange(len(df['Month']))
width = 0.4

plt.bar(x, df['Production_A'], width, color='blue', label='Production A')
plt.bar(x + width, df['Production_B'], width, color='green', label='Production B')

plt.xticks(x + width/2, df['Month'])
plt.xlabel('Month')
plt.ylabel('Production Units')
plt.title('Bar Plot: Production A vs Production B')
plt.legend()
plt.show()

```

```python

box_df = pd.DataFrame({
    "Section_A": [55, 63, 65, 70, 75, 82, 87, 117],
    "Section_B": [29, 50, 59, 62, 65, 70, 79, 83]
})
```

BOX PLOT

```python

box_df.plot(kind="box")
plt.title("Box Plot: Score Distribution")
plt.ylabel("Scores")
plt.show()
```

PIE CHART

```python
region_count = df['Region'].value_counts()

plt.pie(
    region_count.values,
    labels=region_count.index,
    autopct='%1.1f%%',
    colors=['skyblue','lightgreen','salmon','gold'],
    startangle=90
)

plt.title('Pie Chart: Region-wise Distribution')
plt.show()

```

LINE CHART

```python
plt.plot(df['Month'], df['Product_A_Sales'],
         color='blue', marker='o', label='Product A')

plt.plot(df['Month'], df['Product_B_Sales'],
         color='red', marker='s', label='Product B')

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Line Chart: Product A vs Product B Sales')
plt.legend()
plt.show()

```

HEAT MAP

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

numeric_df = df[['Product_A_Sales', 'Product_B_Sales', 'Production_A', 'Production_B']]
corr_matrix = numeric_df.corr()

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    linewidths=0.5
)

plt.title('Heat Map: Correlation Between Sales and Production')
plt.show()
```
